 ## 2. Dictionaries

 Managing four separate variables for a single book can get disorganized. A **dictionary** lets us store related information together using "key-value" pairs.

```python
book_info = {
    "title": book_title,
    "author": book_author,
    "stock": book_stock,
    "price": book_price,
}
```

 ### Think of a dictionary as a folder of labeled files!

 As the illustration below shows, a dictionary groups related data together under one unified name so it is easier to manage.



 - The **dictionary name** (`book_info`) is like the main folder window.

 - The **keys** (`title`, `author`, `stock`, `price`) act like the labels on the files inside.

 - The **values** (`1984`, `George Orwell`, `6`, `14.99`) are the actual pieces of data stored inside each labeled file.



 Instead of remembering *where* a piece of information is stored, you simply ask the dictionary for its specific key, and it hands you the matching value.



 <img src="https://i.imgur.com/sTXGFwW.jpeg" alt="Dictionary Structure" width="400">

```python
# All the information about the book is now stored in a single container. Let's check it out!
print(book_info)
```

```python
# We can pull out specific pieces of information by referencing their "key" in square brackets
book_title = book_info["title"]
print("The novel", book_title, "was written by", book_info["author"])
```

```python
# Let's create two more books to add to our inventory
book_info_2 = {
    "title": "The Little Prince",
    "author": "Antoine de Saint-Exupery",
    "price": 12.49,
    "stock": 17,
}

book_info_3 = {
    "title": "War and Peace",
    "author": "Leo Tolstoy",
    "price": 19.99,
    "stock": 6,
}
```

 ## 3. Lists



 What if we want to store information about multiple books? A **list** allows us to keep an ordered collection of items, which can be of any data type (including dictionaries!).

```python
all_books = [book_info, book_info_2, book_info_3]
```

 The `all_books` variable above is a **list** (indicated by square brackets `[]`).



 **Think of a list as a bookshelf!**

 As shown in the illustration below, a list stores items in a specific, ordered sequence.

 Notice how the bottom shelf is labeled `book 0`? This is a crucial concept: **Python is zero-indexed**. Counting always starts at zero. The first item is at position `0`, the second is at position `1`, and so on.



 <img src="https://i.imgur.com/2pEgvcB.jpeg" alt="List as a Bookshelf" width="300">

```python
print(all_books)
# We can access specific books by their position in the list (starting from 0)
print(all_books[0])  # First book
print(all_books[1])  # Second book
```

 ### Iterating Through a List



 When we have a list of items, we often want to perform the same action on each item. This is where **loops** come in handy. A `for` loop allows us to go through each item in the list one by one.

```python
# Print book titles one by one
for book in all_books:
    print(book["title"])
```

 **Curious how that worked?** Highlight the lines in the above code block, click the Gemini icon, and ask it to explain the syntax.

```python
# Now, let's calculate the total stock of all books.
# We can do this by iterating through the list and summing up the "stock" values for each book.
total_stock = 0  # Initialize total stock to 0
for book in all_books:
    total_stock = total_stock + book["stock"]  # Add the stock of each book to the total

print("Total stock of all books combined:", total_stock)
```

 What if we want to create a list of all book prices? We can use a loop to extract the price from each book dictionary and store it in a new list.

```python
# Start with an empty list to store the prices
book_prices = []
print("Initial list of book prices:", book_prices)

# Iterate through each book and extract the price, then add it to the book_prices list
for book in all_books:
    book_prices.append(book["price"])
    print("List of book prices after adding a new item:", book_prices)

print("Final list of book prices:", book_prices)
```

The `.append()` method adds a single item to the end of a list.

```python
book_prices[0] = 13.99
print(book_prices)
```

```python
# Calculate the total price of all books in one step.
# `book_prices` already contains the price of each book, and `sum()` adds them together.
total_price = sum(book_prices)

print("Total price of all books combined:", total_price)
```

 ### Aggregating Data

 We can easily summarize data to find patterns.

```python
# Calculate the number of book titles for each publisher
books_df["publisher"].value_counts()
```

 **Curious how that worked?** Highlight the line, and ask Gemini it to explain the syntax.

```python
# Show the total number of units sold for each genre using groupby
books_df.groupby("genre")["units_sold"].sum()
```

 The above line performs a two-step aggregation operation: first grouping data, then summing it.


The `.groupby('genre')` method partitions the dataframe into separate groups, where each group contains all rows that share the same value in the `genre` column. Think of it like sorting books into piles by genre: all fiction books together, all children books together, and so on.

The `['units_sold']` selector then picks a single column from each group to work with. This narrows the focus to just the sales figures within each genre group.

Finally, `.sum()` adds up all the values in that column for each group independently. The result is a Series where the index shows each unique genre and the values show the total units sold across all books in that genre. This is a common summarization pattern: group by a category, select a metric, and apply an aggregation function such as `sum`, `mean`, or `count`.

 ### Exploring Relationships with a Correlation Matrix



 Sometimes we want to see how different numerical variables relate to one another. A **correlation matrix** measures these relationships on a scale from -1 to 1:

 * **1** means a perfect positive relationship (as one goes up, the other goes up).

 * **-1** means a perfect negative relationship (as one goes up, the other goes down).

 * **0** means no relationship at all.



 Let's visualize the correlation between a book's average rating, its total number of ratings, and the gross sales using a Seaborn **heatmap**.

```python
# Select the columns we want to compare
cols_to_compare = ["book_average_rating", "book_ratings_count", "gross_sales"]

# Calculate the correlation matrix
corr_matrix = books_df[cols_to_compare].corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1, fmt=".2f")
plt.title("Correlation Matrix: Ratings vs. Sales")
plt.show()
```

 ## 7. Building a Simple Regression Model

Can we predict how many copies a book will sell based on its rating? In this analysis, we'll use **SciPy**, a Python library for scientific and technical computing.

   [<img src="https://scipy.org/images/logo.svg" alt="Dictionary Structure" width="200">](https://scipy.org/)

 Let's use the `linregress()` function from `scipy.stats` to fit a straight line through the data and get its equation.

 The function returns everything we need in one call:
 - **slope** and **intercept** define the line ($y = mx + b$)
 - **r_value** measures how well the line fits (closer to ±1 is better)


```python
from scipy import stats

x = books_df["book_average_rating"]
y = books_df["units_sold"]

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

equation = f"y = {intercept:.2f} + {slope:.2f}x"

print("Equation:", equation)
print("R² =", round(r_value**2, 4))
```

 Let's visualize the model's predictions against the actual data.

```python
# Create points for your line plot
on_the_line = intercept + slope * x

display(on_the_line)
```

*Note:* When you perform arithmetic operations like addition or multiplication with a Pandas Series (like `x`) and scalar (single) numbers (like `intercept` and `slope`), Pandas automatically applies the operation to each element of the Series. This is called vectorization. Because `x` is a Series, the result (`on_the_line`) is also a Series, where each element has been calculated individually.

```python
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color="blue", alpha=0.5, label="Actual Data")
plt.plot(x, on_the_line, color="red", linewidth=2, label=equation)
plt.title("Book Rating vs. Units Sold")
plt.xlabel("Book Rating")
plt.ylabel("Units Sold")
plt.legend()
plt.show()
```

 **Your Turn:** Now let's try a different pair: use `book_ratings_count` to predict `gross_sales`. We'll also use the resulting equation to estimate sales for a specific number of ratings.


```python
x = books_df["book_ratings_count"]
y = books_df["gross_sales"]

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

equation = f"y = {intercept:.2f} + {slope:.2f}x"

print("Equation:", equation)
print("R² =", round(r_value**2, 4))
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color="blue", alpha=0.5, label="Actual Data")
plt.plot(x, slope * x + intercept, color="red", linewidth=2, label=equation)
plt.title("Ratings Count vs. Gross Sales")
plt.xlabel("Total Number of Ratings")
plt.ylabel("Gross Sales ($)")
plt.legend()
plt.show()
```

 Now, let's use our equation directly to estimate gross sales for a book with 167,500 ratings. No model object needed — just plug the number in!


```python
# Plug a new value directly into our equation: y = intercept + slope * x
ratings_count = 167500
predicted_sales = intercept + slope * ratings_count

print("Predicted gross sales for", ratings_count, "ratings: $", round(predicted_sales, 2))
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color="blue", alpha=0.3, label="Historical Data")
plt.plot(x, slope * x + intercept, color="red", linewidth=2, label=equation)
plt.scatter(ratings_count, predicted_sales, color="green", s=100, zorder=5, label="Prediction")

plt.title("Ratings Count vs. Gross Sales (With New Prediction)")
plt.xlabel("Total Number of Ratings")
plt.ylabel("Gross Sales ($)")
plt.legend()
plt.show()
```
